```python
# Import dependencies
import pandas as pd
```

```python
# Read csv into dataframe
aqi = pd.read_csv('../Data/Pollution_day_2001-2016.csv')
aqi.head(5)
```

```python
aqi.dtypes
```

```python
# Filter data by California
aqi_cal = aqi.loc[aqi['State'] == 'California']
aqi_cal
```

```python
# Print county names
aqi_cal['County'].unique().tolist()
```

```python
# Select only years 2013 to 2016
aqi_cal_2009 = aqi_cal.loc[(aqi_cal['Date Local'].str.contains('2009'))]
aqi_cal_2012 = aqi_cal.loc[(aqi_cal['Date Local'].str.contains('2012'))]
aqi_cal_2013 = aqi_cal.loc[(aqi_cal['Date Local'].str.contains('2013'))]
aqi_cal_2014 = aqi_cal.loc[(aqi_cal['Date Local'].str.contains('2014'))]
aqi_cal_2015 = aqi_cal.loc[(aqi_cal['Date Local'].str.contains('2015'))]
aqi_cal_2016 = aqi_cal.loc[(aqi_cal['Date Local'].str.contains('2016'))]
```

```python
# Merge data into one dataframe
#start with merging two data, then continue adding
aqi_data = aqi_cal_2013.merge(aqi_cal_2014, how='outer')
aqi_data = aqi_data.merge(aqi_cal_2012, how='outer')
aqi_data = aqi_data.merge(aqi_cal_2009, how='outer')
aqi_data = aqi_data.merge(aqi_cal_2015, how='outer')
aqi_data = aqi_data.merge(aqi_cal_2016, how='outer')
```

```python
aqi_data
```

```python
# Save dataframe to csv file
aqi_data.to_csv('../Data/aqi_data.csv', index=False)
```
