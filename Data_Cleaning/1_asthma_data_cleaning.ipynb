```python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd

#load data frames from the Data folder
df_asthma = pd.read_csv('../Data/Asthma_Emergency_Department_Visit_Rates_by_ZIP_Code.csv')
df = pd.read_csv('../Data/asthma_data.csv')
```

```python
#replace the two age groups with binary numbers for df emergency visit dataframe
df['Age_Group'].replace(('0-17', '18+'), (1,0), inplace=True)
#Convert County values of df to lower case to match with df_asthma_v1 (later)
df['county'] = df['county'].str.title()
df['county'] = df['county'].str.upper()
df
```

```python
df_asthma
#df_asthma.dtypes
```

```python
#drop all AllAges or All Ages values in Age group column of df_asthma. We only want data for different age groups.

df_asthma_v1 = df_asthma[(df_asthma['Age Group'] != 'AllAges') & (df_asthma['Age Group'] != 'All Ages')]
df_asthma_v1
```

```python
#convert nans to 0
df_asthma_v1 = df_asthma_v1.fillna(0)
df_asthma_v1
```

```python
#Split ZIP code into lat_Lon and then split lat_Lon into Lat and Lon
df_asthma_v1[['ZIP code', 'Lat_Lon']] = df_asthma_v1['ZIP code'].str.split('\n', 1, expand=True)

df_asthma_v1[['Lat', 'Lon']] = df_asthma_v1['Lat_Lon'].str.split(',', 1, expand=True)

df_asthma_v1.drop(columns=['Lat_Lon'],inplace=True)
df_asthma_v1
```

```python
#Remove paranthesis from lat and lon, this is only for learning, lat and Lon will not be used in this analysis
df_asthma_v1['Lat'] = df_asthma_v1['Lat'].str[1:]
df_asthma_v1['Lon'] = df_asthma_v1['Lon'].str[:-1]
#remove all unnecessary columns
df_asthma_v1 = df_asthma_v1.drop(columns={'County Fips code', 'Lat','Lon'})
df_asthma_v1
```

```python
df_asthma_v1 = df_asthma_v1.rename(columns={'ZIP code': 'Zip_Code','Age Group':'Age_Group'})
df_asthma_v1
```

```python
#Encode Age group into binary values because there are only two groups
df_asthma_v1['Age_Group'].replace(('Children (0-17)', 'Adults (18+)'), (1, 0), inplace=True)
df_asthma_v1
```

```python
#compare years of df_asthma_v1 and df
df_asthma_v1['Year'].unique()
```

```python
#make the column names of df consistent with that of df_asthma_v1
df = df.rename(columns={'Number_of_Asthma_ED_Visits': 'Number of Visits','Age_Adjusted_Rate_of_Asthma_ED_V':'Age-adjusted rate','county':'County'})
```

```python
df_asthma_v1['Zip_Code'] = pd.to_numeric(df_asthma_v1['Zip_Code'])
```

```python
df['Number of Visits'] = df['Number of Visits'].astype(float)
```

```python
#merge both data frames into one
asthma_df = df.merge(df_asthma_v1, how='outer', on=['Year','Zip_Code','Age_Group','Number of Visits','Age-adjusted rate', 'County'])
```
