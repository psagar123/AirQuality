```python
import pandas as pd
```

```python
aqi = pd.read_csv('../Data/aqi_data.csv')
asthma = pd.read_csv('../Data/asthma2009_2018.csv')
```

```python
#first work on asthma data

asthma.columns.tolist()
```

```python
asthma
```

```python
#drop zipcode
asthma_clean = asthma.drop(['Zip_Code'], axis=1).copy()
```

```python
asthma_clean
```

```python
#calculate average by county
asth = asthma_clean.groupby(['Year','Age_Group',pd.Grouper(key='County')]).mean()
asth
```

```python
asthma_to_merg = asth.reset_index()
#When we reset the index, the old index (Year) is added as a column, and a new sequential index (0,1....) is used:


asthma_to_merg
```

```python
#now work on aqi data

aqi
```

```python
len(asthma_clean['County'].unique().tolist())
```

```python
#add Year as a separate column
aqi['Year'] = aqi['Date Local'].str.split('-').str[0]
```

```python
#convert the Year string to numeric
aqi['Year'] = pd.to_numeric(aqi['Year'])
```

```python
#since we have Year now, we can drop Date Local
aqi = aqi.drop('Date Local', axis = 1)
```

```python
aqi.columns
```

```python
aqi_grouped = aqi.groupby('Year')
aqi_grouped
```

```python
aqi_grouper = aqi.groupby(['Year',pd.Grouper(key='County')]).agg(['mean','max','min'])
```

```python
aqi_grouper
```

```python
#again reset index so that Year is added as a column and new index is created
aqi_cleaned = aqi_grouper.reset_index()
aqi_cleaned
```

```python
aqi_cleaned.columns = aqi_cleaned.columns.map('_'.join)
aqi_cleaned.columns
```

```python
aqi_cleaned.rename(columns={'Year_': 'Year',
                           'County_':'County'}, inplace = True)
aqi_cleaned
```

```python
#capitalize the first letter of the second part of the county name
aqi_cleaned['County'] = aqi_cleaned['County'].str.title()

aqi_cleaned['County'] = aqi_cleaned['County'].str.upper()
aqi_cleaned

```

```python
asthma_to_merg['Year'].dtype
```

```python
merged = asthma_to_merg.merge(aqi_cleaned, how='inner', on=['County','Year'])
```

```python
print(f'rows: {len(merged)}, columns: {len(merged.columns.tolist())}')
```

```python
merged
```

```python
merged.to_csv('../Data/aqi_asthma_merged.csv', index=False)
```
